# RTBM sampling example

```python
from theta import rtbm
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

We first create a RTBM:

```python
P = rtbm.RTBM(1,1)

P.q = np.diag([110])
P.t = np.diag([1])

P.bv = np.zeros((1,1))
P.bv[0,0] = 0

P.bh = np.zeros((1,1))
P.bh[0,0] = 4

P.w = np.zeros((1,1))
P.w[0,0] = 10
```

Now we perform a sampling with 5000 items.

```python
%time sampling, _ = P.make_sample(5000)
```

Finally, we plot some results:

```python
X = np.linspace(-10,20, 101)
X = X.reshape((1,X.shape[0]))
plt.figure()
plt.xlabel("$v$")
plt.ylabel('$P$', rotation=0, labelpad=10)
plt.plot(X.flatten(), P(X).flatten(), label='RTBM')
hist = plt.hist(sampling, density=True, bins=50, label='sampling')
plt.title('sampling with rotation N=5000')
plt.legend()
```
